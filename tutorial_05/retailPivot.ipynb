```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
import pyspark.sql.functions as f
```

```python
from pyspark.sql import SparkSession

spark= SparkSession.builder.appName('dataframe app').getOrCreate()
```

```python
# load the retail dataset:
retail_data=spark.read.csv("online-retail-dataset.csv",inferSchema=True,header=True)
retail_data.show()
```

```python
result = retail_data.groupBy("StockCode").pivot("Country").sum("Quantity").orderBy("StockCode")
result = result.na.fill(0)
result.show()
```
